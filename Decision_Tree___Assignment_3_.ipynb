# Question 1:  What is a Decision Tree, and how does it work in the context of classification?

### A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It works like a flowchart structure
# where:

- Each internal node represents a decision (condition on a feature)
- Each branch represents the outcome of that decision
- Each leaf node represents the final prediction (class label)

# How it works in Classification:
In classification, a Decision Tree divides the dataset into smaller subsets based on feature values. It selects the best feature to split the data at each step using impurity measures like:

- Gini Impurity
- Entropy (Information Gain)

# Steps involved:
1. Select the best feature based on impurity measure
2. Split the dataset into subsets
3. Repeat the process for each subset
# Stop when:
- All data points belong to the same class OR
- Maximum depth is reached

# Example:

: Suppose we want to classify whether an email is Spam or Not Spam.

- If email contains "Free" → Spam

- Else check another condition → Not Spam

# Advantages:

Easy to understand and interpret

Requires less data preprocessing

Works well with both numerical and categorical data

# Disadvantages:

Can overfit the data

Sensitive to small changes in data


# Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures. How do they impact the splits in a Decision Tree?

# What are Impurity Measures?

Impurity measures are used in Decision Trees to determine how mixed the data is in a node. The goal is to split the data in such a way that each node becomes as pure as possible (i.e., contains data points of only one class).

# Gini Impurity:

Gini Impurity measures the probability of incorrectly classifying a randomly chosen element if it is randomly labeled according to the class distribution.

Decision Tree tries different splits and calculates impurity for each.

# Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each.

# What is Pruning?

Pruning is a technique used in Decision Trees to reduce overfitting by removing unnecessary branches and making the model simpler and more generalizable.

# Pre-Pruning (Early Stopping):

Pre-Pruning stops the growth of the tree before it becomes too complex.

# Conditions used:

max_depth (maximum depth of tree)

min_samples_split

min_samples_leaf

The tree stops splitting when these conditions are met.

# Advantage of Pre-Pruning:

Faster training because the tree stops early

Reduces complexity from the beginning

Post-Pruning (Backward Pruning):

Post-Pruning allows the tree to grow fully and then removes unnecessary branches afterward.

# Process:

1. Build full tree
2. Remove branches that do not improve performance
3. Evaluate using validation data

# Advantage of Post-Pruning:

Produces better accuracy because it considers the full tree before simplifying

More reliable than pre-pruning


```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Pre-Pruning (limit depth)
model_pre = DecisionTreeClassifier(max_depth=3)
model_pre.fit(X_train, y_train)
y_pred_pre = model_pre.predict(X_test)

# Post-Pruning using ccp_alpha
model_post = DecisionTreeClassifier(ccp_alpha=0.01)
model_post.fit(X_train, y_train)
y_pred_post = model_post.predict(X_test)

# Accuracy comparison
print("Accuracy with Pre-Pruning:", metrics.accuracy_score(y_test, y_pred_pre))
print("Accuracy with Post-Pruning:", metrics.accuracy_score(y_test, y_pred_post))
```

```python
# # Question 6:   Write a Python program to:
# ● Load the Iris Dataset
# ● Train a Decision Tree Classifier using the Gini criterion
# ● Print the model’s accuracy and feature importances
# Import libraries

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Load Iris dataset
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names

# Spliting tha  dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create Decision Tree model using Gini
model = DecisionTreeClassifier(criterion='gini')

# Train model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Feature importances
print("\nFeature Importances:")
for name, importance in zip(feature_names, model.feature_importances_):
    print(f"{name}: {importance}")
```

```python
# # Question 7:  Write a Python program to:
# ● Load the Iris Dataset
# ● Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to
# a fully-grown tree.

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics


data = load_iris()
X = data.data
y = data.target

# Spliting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Fully-grown tree
model_full = DecisionTreeClassifier()
model_full.fit(X_train, y_train)
y_pred_full = model_full.predict(X_test)

# Model 2: Limited depth tree
model_limited = DecisionTreeClassifier(max_depth=3)
model_limited.fit(X_train, y_train)
y_pred_limited = model_limited.predict(X_test)

# Accuracy comparison
acc_full = metrics.accuracy_score(y_test, y_pred_full)
acc_limited = metrics.accuracy_score(y_test, y_pred_limited)

print("Accuracy of Fully-Grown Tree:", acc_full)
print("Accuracy of max_depth=3 Tree:", acc_limited)
```

```python
# # Question 8: Write a Python program to:
# ● Load the California Housing dataset from sklearn
# ● Train a Decision Tree Regressor
# ● Print the Mean Squared Error (MSE) and feature importances

import numpy
import pandas
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

# Load dataset
data = fetch_california_housing()
X = data.data
y = data.target
feature_names = data.feature_names

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create Decision Tree Regressor
model = DecisionTreeRegressor()

# Train model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate MSE
mse = metrics.mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

# Feature importances
print("\nFeature Importances:")
for name, importance in zip(feature_names, model.feature_importances_):
    print(f"{name}: {importance}")
```

```python
# # Question 9: Write a Python program to:
# ● Load the Iris Dataset
# ● Tune the Decision Tree’s max_depth and min_samples_split using
# GridSearchCV
# ● Print the best parameters and the resulting model accuracy

import numpy
import pandas
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define model
model = DecisionTreeClassifier()

# Define parameter grid
param_grid = {
    'max_depth': [2, 3, 4, 5, None],
    'min_samples_split': [2, 5, 10]
}

# Apply GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best model
best_model = grid_search.best_estimator_

# Predictions
y_pred = best_model.predict(X_test)

# Accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Model Accuracy after tuning:", accuracy)
```

# Step 1: Handling Missing Values

In real-world healthcare datasets, missing values are common.

# Techniques:

Numerical features → Fill with mean/median

Categorical features → Fill with most frequent value (mode)

Or remove rows/columns if too many missing values

# Example:

Missing age → replace with average age

Missing gender → replace with most common value

🔹 Step 2: Encoding Categorical Features

Machine learning models cannot understand text data directly.

# Techniques:

Label Encoding → Convert categories into numbers (e.g., Male=0, Female=1)

One-Hot Encoding → Create separate columns for each category

# Example:

Disease type → [Diabetes, Cancer] → converted into numeric form

3. Step 3: Training a Decision Tree Model

Split data into training and testing sets

Train a Decision Tree Classifier on training data

min_samples_split → minimum samples to split
