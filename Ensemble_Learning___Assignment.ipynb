# Question 1:  What is Ensemble Learning in machine learning? Explain the key idea behind it.

# Answer:

Ensemble Learning is a machine learning technique in which multiple models are combined to produce a better and more accurate prediction than a single model.

Instead of relying on one model, ensemble learning uses a group of models (called base learners) and combines their predictions.

# Key Idea

> “Many weak learners together become a strong learner.”

Each individual model may make mistakes, but when their predictions are combined, the errors cancel out and performance improves.

# Why Ensemble Learning works
- Reduces overfitting
- Improves accuracy
- Reduces variance and bias
- Produces more stable predictions
# Common Ensemble Methods
- Bagging (Random Forest)
- Boosting (AdaBoost, Gradient Boosting)
- Voting / Stacking




# Question 2: What is the difference between Bagging and Boosting?
# Answer:
| Feature          | **Bagging**                      | **Boosting**                       |
| ---------------- | -------------------------------- | ---------------------------------- |
| Full Form        | Bootstrap Aggregating            | Boosting                           |
| Goal             | Reduce **variance**              | Reduce **bias + variance**         |
| Training style   | Models trained **independently** | Models trained **sequentially**    |
| Data sampling    | Random sampling with replacement | Focus on misclassified samples     |
| Weight of models | Equal weight                     | Different weight                   |
| Overfitting      | Reduces overfitting              | Can overfit if too many iterations |
| Example          | Random Forest                    | AdaBoost, XGBoost                  |
# Summary
Bagging: Parallel learning
Boosting: Sequential learning (learn from mistakes)




# Question 3: What is bootstrap sampling and what role does it play in Bagging methods like Random Forest?

Bootstrap Sampling is a sampling technique in which we randomly select samples from a dataset with replacement.

# With replacement means:

- Same data point can appear multiple times
- Some data points may not appear at all
# Example

Original dataset size = 100
Bootstrap sample = 100 samples drawn randomly with replacement.

# Role in Bagging / Random Forest

In Bagging:

- Multiple datasets are created using bootstrap sampling.
- Each dataset trains a different model.
- Predictions are combined (majority voting or averaging).
# Why it is important
- Creates diverse models
- Reduces overfitting
- Improves generalization

# Question 4: What are Out-of-Bag (OOB) samples and how is OOB score used to evaluate ensemble models?
# Answer:

When bootstrap sampling is used, not all data points are selected for training a model.

The data points not included in a bootstrap sample are called:

 Out-of-Bag (OOB) samples

Important Fact

On average:

- ~63% data → used for training
- ~37% data → becomes OOB samples
# How OOB Score Works

1. Each tree in Random Forest is trained on a bootstrap sample.
1. he remaining OOB samples are used as test data for that tree.
2. Predictions from all trees are combined.
3. Accuracy on OOB samples = OOB Score
# Why OOB Score is useful
- No need for separate test dataset
- Provides internal model validation
- Saves training time


# Question 5: Compare feature importance analysis in a single Decision Tree vs. a Random Forest.
# Answer:

| Feature     | **Decision Tree**                  | **Random Forest**     |
| ----------- | ---------------------------------- | --------------------- |
| Based on    | Single tree splits                 | Average of many trees |
| Stability   | Unstable                           | Very stable           |
| Overfitting | High risk                          | Low risk              |
| Reliability | Less reliable                      | More reliable         |
| Bias        | Can be biased to specific features | Balanced and robust   |

# Explanation
- In a Decision Tree, feature importance depends on one tree → may be biased.
- In a Random Forest, importance is averaged across many trees → more accurate.









```python
# Question 6: Write a Python program to:
# ● Load the Breast Cancer dataset using
# sklearn.datasets.load_breast_cancer()
# ● Train a Random Forest Classifier
# ● Print the top 5 most important features based on feature importance scores.
# (Include your Python code and output in the code box below.)
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# load data
data = load_breast_cancer()

# train model
model = RandomForestClassifier()
model.fit(data.data, data.target)

# feature importance
imp = pd.Series(model.feature_importances_, index=data.feature_names)

# print top 5 features
print(imp.sort_values(ascending=False).head(5))
```

```python
# Question 7: Write a Python program to:
# ● Train a Bagging Classifier using Decision Trees on the Iris dataset
# ● Evaluate its accuracy and compare with a single Decision Tree
# (Include your Python code and output in the code box below.)


from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42)

# Single Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)

# Bagging with Decision Trees
bag = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)
bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)
bag_acc = accuracy_score(y_test, bag_pred)

print("Decision Tree Accuracy:", dt_acc)
print("Bagging Accuracy:", bag_acc)
```

```python
# Question 8: Write a Python program to:
# ● Train a Random Forest Classifier
# ● Tune hyperparameters max_depth and n_estimators using GridSearchCV
# ● Print the best parameters and final accuracy
# (Include your Python code and output in the code box below.)


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# load data
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42)

# model
rf = RandomForestClassifier(random_state=42)

# grid parameters
params = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 10]
}

# grid search
grid = GridSearchCV(rf, params, cv=5)
grid.fit(X_train, y_train)

# best model
best_model = grid.best_estimator_
pred = best_model.predict(X_test)

print("Best Parameters:", grid.best_params_)
print("Final Accuracy:", accuracy_score(y_test, pred))
```

```python
# Question 9: Write a Python program to:
# ● Train a Bagging Regressor and a Random Forest Regressor on the California
# Housing dataset
# ● Compare their Mean Squared Errors (MSE)
# (Include your Python code and output in the code box below.)



from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor

# load data
data = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42)

# Bagging Regressor
bag = BaggingRegressor(estimator=DecisionTreeRegressor(), n_estimators=50, random_state=42)
bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)
bag_mse = mean_squared_error(y_test, bag_pred)

# Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_pred)

print("Bagging MSE:", bag_mse)
print("Random Forest MSE:", rf_mse)
```

# - K-Fold Cross Validation
# - Hyperparameter tuning (GridSearchCV)
# - Limiting tree depth (max_depth)
# - Regularization and early stopping
# These techniques help the model perform well on new unseen data.

# 3. Selecting Base Models

# - Handle large datasets easily
# - Capture complex relationships between features

# 4. Evaluating Performance using Cross-Validation
# Use K-Fold Cross Validation (k = 5 or 10):
# Steps:
# 1. Split dataset into k equal parts
# 2. Train model on k-1 parts and test on remaining part
# 3. Repeat k times and take average performance

# Evaluation Metrics:
# - Accuracy
# - Precision
# - Recall
# - F1 Score
# - ROC-AUC (most important because loan datasets are usually imbalanced)

# 5. Why Ensemble Learning Improves Decision-Making

# - Increasing prediction accuracy
# - Reducing risk of approving loans to defaulters
# - Providing stable and reliable predictions
# - Helping banks reduce financial losses
# - Improving overall credit risk assessment
