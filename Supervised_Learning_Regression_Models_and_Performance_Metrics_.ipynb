# Question 2: What are the key assumptions of Simple Linear Regression?

Simple Linear Regression works correctly only when certain assumptions are satisfied.

#  Key Assumptions:
1. Linearity

There must be a linear relationship between the independent variable (X) and dependent variable (Y).

2. Independence of Errors

The residuals (errors) should be independent of each other.

3. Homoscedasticity (Constant Variance)

The variance of errors should be constant across all values of X.

4. Normality of Errors

The residuals should follow a normal distribution.

5. No Multicollinearity

In Simple Linear Regression, this is automatically satisfied since there is only one independent variable.

# Question 6: What is Logistic Regression? How does it differ from Linear Regression?

# Logistic Regression is a classification algorithm used to predict binary outcomes (0 or 1).
# example
- Spam vs Not Spam
- Pass vs Fail
- Disease vs No Disease

# Concept:
- Uses the sigmoid function
- Output lies between 0 and 1
- Represents probability
# Difference: Logistic vs Linear Regression

| Feature  | Linear Regression | Logistic Regression |
| -------- | ----------------- | ------------------- |
| Output   | Continuous value  | Probability (0–1)   |
| Use Case | Prediction        | Classification      |
| Graph    | Straight Line     | S-shaped (Sigmoid)  |
| Formula  | (Y = β₀ + β₁X)    | Sigmoid function    |
| Example  | House price       | Spam detection      |

# Question 7: Name and briefly describe three common evaluation metrics for regression models.

1. Mean Absolute Error (MAE)
- Measures the average absolute difference between actual and predicted values
 2. Mean Squared Error (MSE)

- Squares errors before averaging
- Penalizes large errors more

 3. Root Mean Squared Error (RMSE)

- Square root of MSE
- Same units as original data

# short ⁉

- MAE → simple and interpretable

- MSE → sensitive to large errors
- RMSE → practical and widely used
# Question 8: What is the purpose of the R-squared metric in regression analysis?
- R-squared is a statistical metric that shows how well the model explains the variability in the data.
-  It measures the proportion of variance in the dependent variable explained by the model.

- R² represents the explained variance ratio

# Range:

R² = 1 → Perfect fit

R² = 0 → No explanatory power

Between 0 and 1 → Partial explanation

# Purpose:
- To evaluate model performance
- To measure goodness of fit
- To compare regression models


```python
# # Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

# Create model
model = LinearRegression()

# Train model
model.fit(X, Y)

# Print slope and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```
