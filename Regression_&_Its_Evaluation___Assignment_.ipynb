# Question 1:  What is Simple Linear Regression?
### Simple Linear Regression is a statistical and machine learning technique used to model the relationship between two variables.

- One variable is the independent variable (X)
- The other is the dependent variable (Y)
### The goal is to predict the value of Y using X.

- It fits a straight line (best-fit line) through the data points to represent the relationship between the variables.

# Key Features
- Uses one independent variable
- Relationship is linear
- Used for prediction and trend analysis
- One of the basic algorithms in machine learning regression


# Question 2: What are the key assumptions of Simple Linear Regression?
# Simple Linear Regression relies on several assumptions

1. Linearity

- There should be a linear relationship between the independent variable (X) and the dependent variable (Y).

- This means the relationship should follow a straight-line pattern.

### Example:
- As study hours increase, exam marks increase approximately linearly.

2. Independence of Errors
- The errors (residuals) should be independent of each other.
- This means the error of one observation should not influence another observation.
3. Homoscedasticity

The variance of the residuals should remain constant across all levels of the independent variable.

In other words, the spread of errors should remain roughly the same for all values of X.

If the spread changes, it is called heteroscedasticity.

4. Normality of Errors

The residuals should follow a normal distribution (bell-shaped curve).

This assumption is particularly important for statistical tests and confidence intervals.

5. No Perfect Multicollinearity

In Simple Linear Regression, there is usually only one independent variable, so multicollinearity is not an issue.

However, in multiple regression models, predictors should not be highly correlated with each other.


# Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Heteroscedasticity occurs in regression analysis when the variance of the errors (residuals) is not constant across all levels of the independent variable.

In simple terms, the spread of the residuals changes as the value of X changes.

If the residual variance increases or decreases across the dataset, the model exhibits heteroscedasticity.

# Example

Consider a regression model predicting expenditure based on income.

- Low-income individuals may have small variations in spending.

- High-income individuals may have larger variations in spending.

This increasing spread of errors represents heteroscedasticity.
# Why it is important to address heteroscedasticity

1. It reduces the reliability of regression predictions.
2. It produces incorrect standard errors, which affect hypothesis testing and p-values.
3. Confidence intervals become unreliable.
4. The model becomes statistically inefficient.

# Question 4: What is Multiple Linear Regression?
Multiple Linear Regression (MLR) is a statistical and machine learning technique used to predict a dependent variable using two or more independent variables.

Unlike Simple Linear Regression, which uses only one predictor, Multiple Linear Regression uses multiple predictors to model the relationship.

The goal is to understand how several factors simultaneously influence the target variable.

# Key Features
- One dependent variable
- Two or more independent variables
- Assumes a linear relationship
- Widely used for prediction, forecasting, and data analysis

Instead of fitting a straight line, it fits a polynomial curve to the data.

#  Question 6:  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:

● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points.




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create Model
model = LinearRegression()

# Train the Model
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print model parameters
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])

# Plot original data points
plt.scatter(X, Y)

# Plot regression line
plt.plot(X, Y_pred)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")

plt.show()
```

# Question 7: Fit a Multiple Linear Regression model on this sample data:

● Area = [1200, 1500, 1800, 2000]

● Rooms = [2, 3, 3, 4]

● Price = [250000, 300000, 320000, 370000]

Check for multicollinearity using VIF and report the results



```python
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create dataset
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Define variables
X = data[['Area', 'Rooms']]
y = data['Price']

# Train Multiple Linear Regression
model = LinearRegression()
model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Add constant column for VIF calculation
X_const = sm.add_constant(X)

# Compute VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]

print("\nVIF Values:")
print(vif_data)
```

# Question 8:  Implement polynomial regression on the following data:

● X = [1, 2, 3, 4, 5]

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Generate polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Plot original data
plt.scatter(X, Y)

# Plot polynomial curve
plt.plot(X, Y_pred)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")

plt.show()
```

# Question 9: Create a residuals plot for a regression model trained on this data:

● X = [10, 20, 30, 40, 50]

● Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.

# Residual=Actual Value−Predicted Value


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1,1)
Y = np.array([15, 35, 40, 50, 65])

# Train regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Compute residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals)

# Horizontal reference line
plt.axhline(y=0)

plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residual Plot")

plt.show()
```
